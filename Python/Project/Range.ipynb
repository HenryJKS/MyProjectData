```python
for index, row in dataExpress.iterrows():
    cep_inicial = row['CEP_INI']
    cep_final = row['CEP_FINAL']

for index, row in dataShipments.iterrows():
    cep = row['CEPS']
    
    if cep_inicial != '' and cep_final != '':
        cep_inicial = int(cep_inicial)
        cep_final = int(cep_final)
            
    ceps = dataShipments['CEPS'].apply(lambda cep: cep_inicial <= cep <= cep_final).sum()

    if ceps > 0:
        cep_range += 1
```

```python
porcentagem = (cep_range / len(dataShipments)) * 100
```
