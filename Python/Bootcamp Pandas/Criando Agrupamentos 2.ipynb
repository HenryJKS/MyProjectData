## Criando Agrupamento II

```python
import pandas as pd
```

```python
dados = pd.read_csv('Datas/aluguel_residencial.csv', sep=';')
```

```python
dados
```

```python
#Agrupando por Area
area = [40, 100, 15, 50, 70]
select = dados['Area'].isin(area)
dados = dados[select]
```

```python
dados['Area'].drop_duplicates()
```

```python
group_area = dados.groupby('Area')
```

```python
group_area.groups
```

```python
group_area[['Valor', 'Condominio']].mean().round(2)
```

## Estatística Descritivas

```python
group_area['Valor'].describe().round(2)
```

```python
group_area['Valor'].aggregate(['sum', 'mean', 'max', 'min']).rename(columns = {'sum' : 'Soma', 'mean' : 'Média', 'max' : 'Máximo', 'min' : 'Mínimo'}).round(2)
```
