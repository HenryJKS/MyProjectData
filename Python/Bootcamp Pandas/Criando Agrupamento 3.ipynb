## Criando Agrupamento

```python
import pandas as pd
```

```python
dados = pd.read_csv('Datas/aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

```python
#Agrupando por 'Tipo'
group_tipo = dados.groupby('Tipo')
```

```python
group_tipo.groups
```

```python
group_tipo[['Valor']].describe().round(2)
```

```python
group_tipo['Valor'].aggregate(['mean', 'sum', 'count']).rename(columns = {'mean' : 'Média', 'sum' : 'Soma', 'count' : 'Unidade'})
```
